```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
%matplotlib inline

print("Hello World!")

#columns = ['mrp', 'price', 'product_category', 'product_name',
#			'pdp_url', 'description', 'color', 'descr_conc',
#			'brand', 'product_category_gen']

df = pd.read_csv('lingerie.csv')

# drop columns with no name
df.drop(df.filter(regex="Unnamed"),axis=1, inplace=True)

df["product_category"].value_counts().head(10).plot(kind="bar")
plt.xlabel("Lingerie Type")
plt.ylabel("Number")
plt.title("Lingerie by Type")
```
